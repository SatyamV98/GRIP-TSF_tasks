```python
# importing the required libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
print("Libraries imported sucessfully")
```

### Importing the data


```python
# importing the dataframe from the provided link
df = pd.read_csv("http://bit.ly/w-data")
print("Data Imported sucessfully")
df.head(10)
```

```python
# Using 'df.tail()' to display last ten rows of the dataframe
df.tail(10)
```

```python
# Using 'df.info' to print informaton about the dataframe
df.info()
```

```python
# Using 'df.describe()' to print statistical information about the dataframe
df.describe()
```

### Input data visualization

```python
# visualizing dataframe using scatter plot
df.plot(x='Hours', y='Scores', style='o',color='green')  
plt.title('Hours vs Percentage',fontsize=15)  
plt.xlabel('Hours Studied',fontsize=10)  
plt.ylabel('Percentage Score',fontsize=10) 
plt.show()
```

```python
# visualizing with line plot
plt.style.use('ggplot')
df.plot(kind='line')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

### Processing the data

```python
# separating the variables into independent and dependent variables
x = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values
```

```python
# further separating into training and testing sets 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=200)
print(x_train.ndim)
print(x_train.shape)
```

```python
print(x_test.ndim)
print(x_test.shape)
```

```python
print(y_train.ndim)
print(y_train.shape)
```

```python
print(y_test.ndim)
print(y_test.shape)
```

### Model training

```python
# training the model using the training set
reg = LinearRegression()  
reg.fit(x_train, y_train) 

print("Training complete!!")
```

### plotting the line of regression

```python
# creating the line
slope=reg.coef_
intercept=reg.intercept_

line=slope*x + intercept # y = m*x + c 
```

```python
# plotting the line
plt.scatter(x, y)
plt.plot(x,line,color='blue')
plt.title('Linear regression')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')
plt.show()
```

### Making Predictions

```python
# testing data
print(x_test)
# model prediction
y_pred = reg.predict(x_test)
```

```python
df1=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df1
```

```python
# Finding accuracy of the training and testing sets for actual vs predicted results
a=reg.score(x_train,y_train)*100
b=reg.score(x_test,y_test)*100

print("Training Accuracy=",a)
print("Testing Accuracy=",b)
```

### actual vs predicted result comparison

```python
# Plotting the Bar graph to depict the difference between the actual and predicted value
df1.plot(kind='bar',figsize=(10,5))
plt.grid(which='major', linewidth='0.5', color='red')
plt.grid(which='minor', linewidth='0.5', color='blue')
plt.grid()
plt.show()
```

## Testing the model against our own data:

### Prediction of score of a student studying 9.25 hrs/day

```python
hours = 9.25
test = np.array([hours])
test = test.reshape(-1, 1)
own_pred = reg.predict(test)
print("No of Hours studied every day = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

### Evaluating the model

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. Here different errors have been calculated to compare the model performance and predict the accuracy.

```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-2 score:', metrics.r2_score(y_test, y_pred))
```
